- Non linear mapping to transform the data into a higher dimensional space
- Then use PCA to project the data back to the lower dimensional space

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, random_state=123)

plt.scatter(X[y==0, 0], X[y==0,1], color='r', marker='^', alpha=0.5)
plt.scatter(X[y==1, 0], X[y==1,1], color='b', marker='^', alpha=0.5)

plt.tight_layout()
plt.show()
```

__This clearly cannot be linearly separated__

```python
from sklearn.decomposition import PCA
import numpy as np
sk_pca = PCA(n_components=2)

X_spca = sk_pca.fit_transform(X)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(7,3))

ax[0].scatter(X_spca[y==0, 0], X_spca[y==0, 1], color='r',
             marker='^', alpha=0.5)
ax[0].scatter(X_spca[y==1, 0], X_spca[y==1, 1], color='b',
             marker='o', alpha=0.5)
ax[1].scatter(X_spca[y==0, 0], np.zeros((50,1))+0.03, color='r',
             marker='^', alpha=0.5)
ax[1].scatter(X_spca[y==1, 0], np.zeros((50,1))-0.03, color='b',
             marker='o', alpha=0.5)

ax[0].set_xlabel("PC1")
ax[0].set_ylabel("PC2")
ax[1].set_ylim([-1,1])
ax[1].set_yticks([])
ax[1].set_xlabel("PC1")
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import KernelPCA
X, y = make_moons(n_samples=100, random_state=123)

sk_kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_skpca = sk_kpca.fit_transform(X)

plt.scatter(X_skpca[y==0, 0], X_skpca[y==0, 1], color='r',
             marker='^', alpha=0.5)
plt.scatter(X_skpca[y==1, 0], X_skpca[y==1, 1], color='b',
             marker='o', alpha=0.5)

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.tight_layout()
plt.show()
```

### KPCA to separate concentric circles

```python
from sklearn.datasets import make_circles

X, y = make_circles(n_samples=1000,random_state=200,  noise=0.10, factor=0.2)

plt.scatter(X[y==0, 0], X[y==0,1], color='r',  marker='^', alpha=0.5)
plt.scatter(X[y==1, 0], X[y==1,1], color='b',  marker='o', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
s_pca = PCA(n_components=2)
X_spca = s_pca.fit_transform(X)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(7,3))

ax[0].scatter(X_spca[y==0, 0], X_spca[y==0, 1], color='r',
             marker='^', alpha=0.5)
ax[0].scatter(X_spca[y==1, 0], X_spca[y==1, 1], color='b',
             marker='o', alpha=0.5)
ax[1].scatter(X_spca[y==0, 0], np.zeros((500,1))+0.03, color='r',
             marker='^', alpha=0.5)
ax[1].scatter(X_spca[y==1, 0], np.zeros((500,1))-0.03, color='b',
             marker='o', alpha=0.5)

ax[0].set_xlabel("PC1")
ax[0].set_ylabel("PC2")
ax[1].set_ylim([-1,1])
ax[1].set_yticks([])
ax[1].set_xlabel("PC1")
plt.tight_layout()
plt.show()
```
